```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('data/train.csv')
```

```python
X=train.drop('price_range',axis=1)
X.head()
```

```python
y = train['price_range']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
 
model = RandomForestClassifier(random_state = 1, n_estimators=300)

model.fit(X_train, y_train)

acuracia = model.score(X_test, y_test)*100
print("A acuracia é de:", acuracia)
```

```python
from sklearn.dummy import DummyClassifier

dummy_clf = DummyClassifier(strategy="most_frequent")

dummy_clf.fit(X_train, y_train)

acuracia_dummy = dummy_clf.score(X_test, y_test)*100
print("A acuracia do dummy classifier é:", acuracia_dummy)
```
