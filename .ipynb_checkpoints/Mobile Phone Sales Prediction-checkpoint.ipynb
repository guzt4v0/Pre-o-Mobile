```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
X=train.drop('price_range',axis=1)
X
```

```python
y = train['price_range']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
```

```python
seed = 20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state = 1, max_depth=3)

model.fit(X_train, y_train)

predicao = model.predict(X_test)

acuracia = model.score(X_test, y_test)*100
print(acuracia)
```

```python
from sklearn.tree import export_graphviz
import graphviz

features = X.columns
dot_data = export_graphviz(model, out_file = None, feature_names=features, filled=True, rounded=True, class_names=["0", "1", "2", "3"])
grafico = graphviz.Source(dot_data)
grafico
```
